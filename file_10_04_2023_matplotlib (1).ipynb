```python
import matplotlib.pyplot as plt
```

```python
#scatter plot
#2 data one for x axis and one for y -axis
import numpy as np
x =  [1,2,3,4]
y = [4,6,8,10]
z = [2,4,6,8]
plt.scatter(x,y,c="red",s=100,marker="h",alpha=1,edgecolors="green")
plt.plot(x,y,c="green",linewidth=5,linestyle='dotted',label="Line1")
plt.scatter(x,z,c="blue",s=100,marker='*',alpha=1,edgecolors="yellow")
plt.plot(x,z,'y-.',label="Line2")
plt.legend()
plt.grid(axis="x")
for i in range(len(x)):
    plt.text(x[i]+0.1,y[i]-2,(x[i],y[i]))
for i in range(len(x)):
    plt.text(x[i],z[i]-0.5,(x[i],z[i]))
plt.axis([0,5,1,11])
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Scatter Plot")

plt.rcParams["figure.figsize"] = (5,5)
plt.show()
#plt.plot()
```

```python
#bubble chart
year = [2012,2013,2014,2015,2016,2017]
salary = [10,11,13,14,15,17]
pop = [100,200,250,700,550,600]

ax = plt.axes()
plt.scatter(year,salary,s=pop,c=["green","yellow","red","orange","black","blue"])

for i in range(len(year)):
    plt.text(year[i],salary[i]-0.5,(pop[i]))
ax.set_facecolor("yellow")
#plt.plot(year,salary)
plt.show()
```

```python
#bar charts
#compare the marks of the learners

stu = ["Himanshu","Mukta","Niranjan","Ritu","Priyanka","Muskan"]
marks = [80,85,88,89,75,90]

stu_1 = ["Prakash","Kiran","Nikitha","Saurav","Nikhil","Mirza"]
marks_1 = [80,85,88,89,75,90]


plt.subplot(2,1,1)
plt.bar(stu,marks,width=0.5,align="center",color=["green","yellow","red","orange","black","blue"])
plt.xticks(rotation=45)
plt.subplot(2,1,2)
plt.bar(stu_1,marks_1,width=0.5,align="center",color=["green","yellow","red","orange","black","blue"])
plt.xticks(rotation=45)

plt.show()
#plt.bar()
```

```python
#pie charts
#to show the distribution of the data
learners = ["A","B","C","D"]
data = [45,49,51,83]
plt.pie(data,labels=learners,counterclock=False,startangle=90,autopct="%.2f%%",explode=[0,0,0,0.2],labeldistance=0.4,shadow=True)
plt.legend(loc=1,title="Learners Marks Distribution",fontsize=10,bbox_to_anchor=(1.1,1.3))
plt.show()
```
